```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
team_a = [0.5, 0.5, 0.5, 0.4, 0.4]
team_b = [0.5, 0.5, 0.5, 0.5, 0.5]

def get_draw(team):
    return np.sum([np.random.normal(wr, 1) for wr in team])

samples_a = [get_draw(team_a) for i in range(10000)]
samples_b = [get_draw(team_b) for i in range(10000)]

team_a_wins = 0
for i in range(10000):
    if samples_a[i] > samples_b[i]:
        team_a_wins += 1

print(team_a_wins / 10000)

plt.hist(samples_a, bins=100, alpha=0.5, label='Team A')
plt.hist(samples_b, bins=100, alpha=0.5, label='Team B')
plt.legend(loc='upper right')
plt.show()
```
